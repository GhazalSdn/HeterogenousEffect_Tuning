```python
import pandas as pd
from sklearn.model_selection import train_test_split
from CTL.causal_tree_learn import CausalTree
import numpy as np
import pickle
```

```python
print("### Load Data ######################")
data = pd.read_csv('data/preprocessed_app_data.csv')
data
```

```python
# columns = ['spark.executor.cores', 
#            'spark.executor.memory',
#            'spark.default.parallelism', 
#            'spark.serializer', 
#            'spark.task.cpus',
#            'executorDeserializeCpuTime', 
#            'executorRunTime', 
#            'executorCpuTime',
#            'resultSize', 
#            'jvmGcTime', 
#            'Execution_Time']
treatment_col = 'spark.executor.cores'
```

```python
data1 = data.sample(n=len(data), replace=True)
data2 = data.sample(n=len(data), replace=True)
data3 = data.sample(n=len(data), replace=True)
```

```python
data1.describe()
```

```python
data2.describe()
```

```python
y1 = data1['Execution_Time'].values
treat1 = data1[treatment_col].values
# # other configs
drop_cols = ['Execution_Time', #Outcome
             treatment_col, #Treatment
             'executorDeserializeCpuTime', 
             'executorRunTime', 
             'executorCpuTime', 
#              'resultSize', 
             'jvmGcTime' # mediators
            ]
x1 = data1.drop(drop_cols, axis=1).values
columns = data1.drop(drop_cols, axis=1).columns
```

```python
y2 = data2['Execution_Time'].values
treat2 = data2[treatment_col].values
# # other configs
drop_cols = ['Execution_Time', #Outcome
             treatment_col, #Treatment
             'executorDeserializeCpuTime', 
             'executorRunTime', 
             'executorCpuTime', 
#              'resultSize', 
             'jvmGcTime' # mediators
            ]
x2 = data2.drop(drop_cols, axis=1).values
columns = data2.drop(drop_cols, axis=1).columns
```

```python
y3 = data3['Execution_Time'].values
treat3 = data3[treatment_col].values
# # other configs
drop_cols = ['Execution_Time', #Outcome
             treatment_col, #Treatment
            'executorDeserializeCpuTime', 
             'executorRunTime', 
             'executorCpuTime',
#              'resultSize', 
             'jvmGcTime' # mediators
            ]
x3 = data3.drop(drop_cols, axis=1).values
columns = data3.drop(drop_cols, axis=1).columns
```

```python
np.random.seed(0)
```

```python
# x1, x2, y1, y2, treat1, treat2 = train_test_split(x, y, treatment, test_size=0.5, random_state=5)
```

# Part 1

```python
x1_train, x1_test, y1_train, y1_test, treat1_train, treat1_test = train_test_split(x1, y1, treat1, test_size=0.2, random_state=42)
```

```python
# regular CTL
ctl = CausalTree(cont=True)
ctl.fit(x1_train, y1_train, treat1_train)
ctl.prune()
ctl_predict = ctl.predict(x1_test)

# honest CTL
cth = CausalTree(cont=True, honest=True)
cth.fit(x1_train, y1_train, treat1_train)
cth.prune()
cth_predict = cth.predict(x1_test)

# val honest CTL
cthv = CausalTree(cont=True, val_honest=True)
cthv.fit(x1_train, y1_train, treat1_train)
cthv.prune()
cthv_predict = cthv.predict(x1_test)

# adaptive CT
ct_adaptive = CausalTree(weight=0.0, split_size=0.0, cont=True)
ct_adaptive.fit(x1_train, y1_train, treat1_train)
ct_adaptive.prune()
ct_adaptive_predict = ct_adaptive.predict(x1_test)

# honest CT
ct_honest = CausalTree(honest=True, weight=0.0, split_size=0.0, cont=True)
ct_honest.fit(x1_train, y1_train, treat1_train)
ct_honest.prune()
ct_honest_predict = ct_honest.predict(x1_test)


# if you want to plot a tree

# ct_adaptive.plot_tree(features=columns, filename="output/p_otherConfigs_cores/adaptive_ct_otherF", show_effect=True)
# ct_honest.plot_tree(features=columns, filename="output/p_otherConfigs_cores/honest_ct_otherF", show_effect=True)
# ctl.plot_tree(features=columns, filename="output/p_otherConfigs_cores/regular_ctl_otherF", show_effect=True)
# cth.plot_tree(features=columns, filename="output/p_otherConfigs_cores/honest_ctl_otherF", show_effect=True)
# cthv.plot_tree(features=columns, filename="output/p_otherConfigs_cores/valHonest_ctl_otherF", show_effect=True)


ct_adaptive.plot_tree(features=columns, filename="output/robust_p_MedConfigs_cores/1adaptive_ct_otherF", show_effect=True)
ct_honest.plot_tree(features=columns, filename="output/robust_p_MedConfigs_cores/1honest_ct_otherF", show_effect=True)
ctl.plot_tree(features=columns, filename="output/robust_p_MedConfigs_cores/1regular_ctl_otherF", show_effect=True)
cth.plot_tree(features=columns, filename="output/robust_p_MedConfigs_cores/1honest_ctl_otherF", show_effect=True)
cthv.plot_tree(features=columns, filename="output/robust_p_MedConfigs_cores/1valHonest_ctl_otherF", show_effect=True)


# ct_adaptive.plot_tree(features=columns, filename="output/robust_p_otherConfigs_cores/1adaptive_ct_otherF", show_effect=True)
# ct_honest.plot_tree(features=columns, filename="output/robust_p_otherConfigs_cores/1honest_ct_otherF", show_effect=True)
# ctl.plot_tree(features=columns, filename="output/robust_p_otherConfigs_cores/1regular_ctl_otherF", show_effect=True)
# cth.plot_tree(features=columns, filename="output/robust_p_otherConfigs_cores/1honest_ctl_otherF", show_effect=True)
# cthv.plot_tree(features=columns, filename="output/robust_p_otherConfigs_cores/1valHonest_ctl_otherF", show_effect=True)

# ct_adaptive.plot_tree(features=columns, filename="output/otherConfigs_parallelism/parallelism_adaptive_ct_otherF", show_effect=True)
# ct_honest.plot_tree(features=columns, filename="output/otherConfigs_parallelism/parallelism_honest_ct_otherF", show_effect=True)
# ctl.plot_tree(features=columns, filename="output/otherConfigs_parallelism/parallelism_regular_ctl_otherF", show_effect=True)
# cth.plot_tree(features=columns, filename="output/otherConfigs_parallelism/parallelism_honest_ctl_otherF", show_effect=True)
# cthv.plot_tree(features=columns, filename="output/otherConfigs_parallelism/parallelism_valHonest_ctl_otherF", show_effect=True)
```
